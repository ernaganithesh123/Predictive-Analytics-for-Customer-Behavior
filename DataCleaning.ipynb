### Import the required libraries.

```python
import pandas as pd
```

### Load the Dataset called customer_churn_training.csv.

```python
data = pd.read_csv("customer_churn_training.csv")
```

```python
data.shape
```

### The dataset contains the 4,40,833 observation with 12 columns.

```python
data.info()
```

#### Only one observation contains the null value.

```python
data.isna().sum()
```

```python
data[data.isna().any(axis=1)]
```

```python
data = data.dropna()
```

#### The null value was droped because no record was entered.

```python
data.dtypes
```

### -----------------------------------DataTypes convertion-----------------------------------
##### The column Total Spend contains the float values and remaining columns can be converted as integer.

```python
data.columns.to_list()
```

```python
int_col = ['CustomerID','Age','Tenure','Usage Frequency','Support Calls','Payment Delay','Last Interaction','Churn'] 

data[int_col] = data[int_col].astype(int)
```

```python
data.dtypes
```

#### 'CustomerID'---'Age'---'Tenure'---'Usage Frequency'---'Support Calls'---'Payment Delay'---'Last Interaction'---'Churn' This columns are converted to integer.

```python
data.info()
```

```python
data.shape
```

### Now the dataset was cleaned and no null values are present.

#### Check the dataset for duplicated values.

```python
data.duplicated().sum()
```

```python
data.to_csv("customer_churn_cleaned.csv",index=False)
```
